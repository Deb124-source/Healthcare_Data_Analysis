# **Import the necessary libraries**

```python
df['Name']=df['Name'].str.title()
df
```

```python
df['Gender']=df['Gender'].replace({'M':'MALE','F':'FEMALE'})
df
```

```python
df['Date of Admission']=df['Date of Admission'].str.replace('/','-',regex=False)
df['Date of Admission']=df['Date of Admission'].str.replace('.','-',regex=False)
df['Discharge Date']=df['Discharge Date'].str.replace('/','-',regex=False)
df['Discharge Date']=df['Discharge Date'].str.replace('.','-',regex=False)
df
```

```python
df['Doctor']=df['Doctor'].str.replace('#','')
df
```

```python
df['Hospital']=df['Hospital'].str.replace(',','')
df
```

```python
df['Name']=df['Name'].fillna("Unknown")
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

```python
df['Gender'].value_counts().index[0]
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df
```

```python
df['Doctor']=df['Doctor'].fillna('Unknown')
df
```

```python
df['Billing Amount']=df['Billing Amount'].str.replace(' USD','')
df
```

```python
print("Missing values before further cleaning:")
print(df.isna().sum())

# Convert date columns to datetime, coercing errors to NaT
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], errors='coerce')
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], errors='coerce')

# Fill NaT values in date columns with the mode (most frequent date)
df['Date of Admission'].fillna(df['Date of Admission'].mode()[0], inplace=True)
df['Discharge Date'].fillna(df['Discharge Date'].mode()[0], inplace=True)

# Convert 'Billing Amount' to numeric (float)
df['Billing Amount'] = pd.to_numeric(df['Billing Amount'], errors='coerce')

# Fill NaNs in 'Billing Amount' with the mean
df['Billing Amount'].fillna(df['Billing Amount'].mean(), inplace=True)

# Fill NaNs in 'Room Number' with the mode
df['Room Number'].fillna(df['Room Number'].mode()[0], inplace=True)

# Fill NaNs in 'Admission Type' with the mode
df['Admission Type'].fillna(df['Admission Type'].mode()[0], inplace=True)

# Fill NaNs in 'Medication' with the mode
df['Medication'].fillna(df['Medication'].mode()[0], inplace=True)

# Fill NaNs in 'Test Results' with the mode
df['Test Results'].fillna(df['Test Results'].mode()[0], inplace=True)

print("\nMissing values after further cleaning:")
print(df.isna().sum())

print("\nData types after cleaning:")
print(df.info())

print("\nDescriptive statistics for numerical columns:")
print(df.describe())

print("\nValue counts for key categorical columns:")
print("Gender:\n", df['Gender'].value_counts())
print("\nMedical Condition:\n", df['Medical Condition'].value_counts())
print("\nAdmission Type:\n", df['Admission Type'].value_counts())
print("\nTest Results:\n", df['Test Results'].value_counts())
```

```python
# Set up the plotting style
sns.set_style("whitegrid")

# Visualize distributions of numerical features
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(df['Billing Amount'], bins=20, kde=True)
plt.title('Distribution of Billing Amount')

plt.subplot(1, 3, 3)
sns.histplot(df['Room Number'], bins=20, kde=True)
plt.title('Distribution of Room Number')

plt.tight_layout()
plt.show()

# Visualize counts of key categorical features
plt.figure(figsize=(20, 6))

plt.subplot(1, 4, 1)
sns.countplot(data=df, x='Gender', palette='viridis')
plt.title('Count of Gender')

plt.subplot(1, 4, 2)
sns.countplot(data=df, x='Medical Condition', palette='viridis')
plt.title('Count of Medical Condition')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 4, 3)
sns.countplot(data=df, x='Admission Type', palette='viridis')
plt.title('Count of Admission Type')

plt.subplot(1, 4, 4)
sns.countplot(data=df, x='Test Results', palette='viridis')
plt.title('Count of Test Results')

plt.tight_layout()
plt.show()

# Explore the number of unique doctors and hospitals
print(f"\nNumber of unique Doctors: {df['Doctor'].nunique()}")
print(f"Number of unique Hospitals: {df['Hospital'].nunique()}")

# Display top 10 Doctors and Hospitals by frequency
print("\nTop 10 Doctors:\n", df['Doctor'].value_counts().head(10))
print("\nTop 10 Hospitals:\n", df['Hospital'].value_counts().head(10))
```

## Further Data Exploration: Relationships and Trends

### Correlation Matrix for Numerical Features

```python
# Calculate the correlation matrix for numerical columns
numerical_cols = ['Age', 'Billing Amount', 'Room Number']
corr_matrix = df[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

### Numerical Features vs. Categorical Features (Box Plots)

```python
plt.figure(figsize=(18, 10))

plt.subplot(2, 3, 1)
sns.boxplot(data=df, x='Gender', y='Age', palette='pastel')
plt.title('Age Distribution by Gender')

plt.subplot(2, 3, 2)
sns.boxplot(data=df, x='Medical Condition', y='Age', palette='pastel')
plt.title('Age Distribution by Medical Condition')
plt.xticks(rotation=45, ha='right')

plt.subplot(2, 3, 3)
sns.boxplot(data=df, x='Admission Type', y='Age', palette='pastel')
plt.title('Age Distribution by Admission Type')

plt.subplot(2, 3, 4)
sns.boxplot(data=df, x='Gender', y='Billing Amount', palette='muted')
plt.title('Billing Amount by Gender')

plt.subplot(2, 3, 5)
sns.boxplot(data=df, x='Medical Condition', y='Billing Amount', palette='muted')
plt.title('Billing Amount by Medical Condition')
plt.xticks(rotation=45, ha='right')

plt.subplot(2, 3, 6)
sns.boxplot(data=df, x='Admission Type', y='Billing Amount', palette='muted')
plt.title('Billing Amount by Admission Type')

plt.tight_layout()
plt.show()
```

### Time Series Analysis: Admissions and Billing Over Time

```python
# Group by Date of Admission to get daily admissions count
daily_admissions = df['Date of Admission'].value_counts().sort_index()

# Group by Date of Admission to get daily average billing amount
daily_avg_billing = df.groupby('Date of Admission')['Billing Amount'].mean()

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
daily_admissions.plot(kind='line')
plt.title('Daily Number of Admissions Over Time')
plt.xlabel('Date of Admission')
plt.ylabel('Number of Admissions')

plt.subplot(1, 2, 2)
daily_avg_billing.plot(kind='line', color='orange')
plt.title('Daily Average Billing Amount Over Time')
plt.xlabel('Date of Admission')
plt.ylabel('Average Billing Amount')

plt.tight_layout()
plt.show()
```

## Further Analysis: Length of Stay and Billing Anomalies

### Calculate and Visualize Length of Stay

```python
# Calculate 'Length of Stay' in days
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

# Handle cases where discharge date is before admission date (if any)
df['Length of Stay'] = df['Length of Stay'].apply(lambda x: x if x >= 0 else 0)

plt.figure(figsize=(10, 6))
sns.histplot(df['Length of Stay'], bins=30, kde=True)
plt.title('Distribution of Length of Stay (Days)')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Count')
plt.show()

print("\nDescriptive statistics for Length of Stay:")
print(df['Length of Stay'].describe())
```

### Investigate Negative Billing Amounts

```python
# Filter and display rows with negative 'Billing Amount'
negative_billing_df = df[df['Billing Amount'] < 0]

print(f"Number of records with negative billing amount: {len(negative_billing_df)}")
if not negative_billing_df.empty:
    print("\nSample of records with negative billing amount:")
    display(negative_billing_df.head())
else:
    print("No records found with negative billing amount.")

# Optional: Consider dropping or correcting these if they are errors. For now, we just investigate.
```

## Medical Condition vs. Billing, and Top Providers

### Average Billing Amount by Medical Condition

```python
avg_billing_by_condition = df.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_billing_by_condition.index, y=avg_billing_by_condition.values, palette='coolwarm')
plt.title('Average Billing Amount by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAverage Billing Amount by Medical Condition:\n", avg_billing_by_condition)
```

### Average Billing Amount by Insurance Provider

```python
avg_billing_by_insurance = df.groupby('Insurance Provider')['Billing Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_billing_by_insurance.index, y=avg_billing_by_insurance.values, palette='viridis')
plt.title('Average Billing Amount by Insurance Provider')
plt.xlabel('Insurance Provider')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAverage Billing Amount by Insurance Provider:\n", avg_billing_by_insurance)
```

### Top 10 Doctors and Hospitals by Total Billing Amount

```python
top_doctors_by_billing = df.groupby('Doctor')['Billing Amount'].sum().sort_values(ascending=False).head(10)
top_hospitals_by_billing = df.groupby('Hospital')['Billing Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=top_doctors_by_billing.index, y=top_doctors_by_billing.values, palette='magma')
plt.title('Top 10 Doctors by Total Billing Amount')
plt.xlabel('Doctor')
plt.ylabel('Total Billing Amount')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
sns.barplot(x=top_hospitals_by_billing.index, y=top_hospitals_by_billing.values, palette='magma')
plt.title('Top 10 Hospitals by Total Billing Amount')
plt.xlabel('Hospital')
plt.ylabel('Total Billing Amount')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("\nTop 10 Doctors by Total Billing Amount:\n", top_doctors_by_billing)
print("\nTop 10 Hospitals by Total Billing Amount:\n", top_hospitals_by_billing)
```
